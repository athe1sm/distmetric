## Sumstat demo
Goal: Demonstrate functionality of SumStat class object

### 1. Import package, generate random trees and calculate quartets distances

```python
import distmetric
```

```python
# generate 10 random trees with Generator and save to variable named randomtrees
TREES = distmetric.Generator(ntrees=20, ntips=10, treeheight=1.0)
randomtrees = TREES.get_randomtrees()

# calculate quartet distances with 20 random trees and PAIRWISE sampling
quart = distmetric.Quartets(trees=randomtrees, sampmethod="pairwise")
quart.run()
df = quart.output
```

```python
df.head()
```

### 2. Summarize quartet distances with SumStat

```python
# using data frame from above, and define column of interest as "Quartet_intersection"
stats = distmetric.SumStat(df, "Quartet_intersection")
```

```python
stats.get_mean()
```

```python
stats.get_std()
```

```python
stats.histogram()
```
